# Concatenate segmented catalogs downloaded from ANSS
2 sets of 6 months' worth of data were downloaded each year from 2010 to 2015 since the attempt to download the data for a whole year will return an error for the latitude and longitude setting for this research

Sources:
 - https://earthquake.usgs.gov/earthquakes/search/
 - https://earthquake.usgs.gov/fdsnws/event/1/query.csv?starttime=2015-1-1%2000:00:00&endtime=2016-1-1%2000:00:00&maxlatitude=51&minlatitude=38&maxlongitude=-119&minlongitude=-132&minmagnitude=0&maxmagnitude=10&eventtype=earthquake&orderby=time-asc
 - https://chatgpt.com/c/6716aa0a-47cc-800c-8385-916b2fbbdfa1

```python
import pandas as pd
```

```python
anss_2010_1_6 = pd.read_csv('../data/datasets_anss/anss_2010-1-6.csv')
anss_2010_7_12 = pd.read_csv('../data/datasets_anss/anss_2010-7-12.csv')
anss_2011_1_6 = pd.read_csv('../data/datasets_anss/anss_2011-1-6.csv')
anss_2011_7_12 = pd.read_csv('../data/datasets_anss/anss_2011-7-12.csv')
anss_2012_1_6 = pd.read_csv('../data/datasets_anss/anss_2012-1-6.csv')
anss_2012_7_12 = pd.read_csv('../data/datasets_anss/anss_2012-7-12.csv')
anss_2013_1_6 = pd.read_csv('../data/datasets_anss/anss_2013-1-6.csv')
anss_2013_7_12 = pd.read_csv('../data/datasets_anss/anss_2013-7-12.csv')
anss_2014_1_6 = pd.read_csv('../data/datasets_anss/anss_2014-1-6.csv')
anss_2014_7_12 = pd.read_csv('../data/datasets_anss/anss_2014-7-12.csv')
anss_2015_1_6 = pd.read_csv('../data/datasets_anss/anss_2015-1-6.csv')
anss_2015_7_12 = pd.read_csv('../data/datasets_anss/anss_2015-7-12.csv')
```

```python
anss_2011_2015 = pd.concat([anss_2010_1_6, anss_2010_7_12,anss_2011_1_6, anss_2011_7_12,anss_2012_1_6,anss_2012_7_12,anss_2013_1_6,
                            anss_2013_7_12,anss_2014_1_6,anss_2014_7_12,anss_2015_1_6,anss_2015_7_12], axis=0, ignore_index=True)
anss_2011_2015.to_csv('../data/datasets_anss/anss_2010-15.csv')
```
