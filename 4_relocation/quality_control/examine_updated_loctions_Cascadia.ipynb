##### Populate the columns of the updated file

```python
df_updated[' Event ID '] = df_ver3[' Event ID ']
df_updated[' Depth (km)'] = df_updated[' Depth (m)']/1000
df_updated[' Detection Value'] = df_ver3[' Detection Value']
df_updated.head()
```

```python
# Convert seconds to HH:MM:SS format
df_updated[' Origin Time (UTC)'] = pd.to_datetime(df_ver3[' Origin Time (UTC)']).dt.date.astype(str) + 'T' +pd.to_datetime(df_updated['sec_day'], unit='s').dt.strftime('%H:%M:%S.%fZ').astype(str)
df_updated.iloc[0:10]
```

```python
df_updated[' RMS Residual (s)'] = df_ver3[' RMS Residual (s)']
df_updated[' Num. P'] = df_ver3[' Num. P']
df_updated[' Num. S'] = df_ver3[' Num. S']
df_updated[' Depth (km)'] = df_updated[' Depth (km)'] * (-1)
```

```python
df_updated = df_updated.drop(columns=[' Depth (m)', 'sec_day'])
df_updated 
```

```python
df_updated.to_csv('/wd1/hbito_data/data/datasets_all_regions/Cascadia_updated_catalog_ver3.csv', index=False)
```
