```python
import pandas as pd
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import table 
import df2img

```

```python
df_arrival = pd.read_csv('/wd1/hbito_data/data/datasets_all_regions/arrival_2010_2015_reloc_cog_ver3.csv',index_col=0)
df_arrival.head()
```

```python
fig = df2img.plot_dataframe(
    df_arrival.head(),

    tbl_header=dict(
        align="center",
        fill_color="purple",
        font_color="white",
        font_size=80,
        line_color="darkslategray",
        height=120
    ),
    tbl_cells=dict(
        align="center",
        line_color="darkslategray",
        font_size=90,
        height=120
    ),
    row_fill_color=("#ffffff", "#d7d8d6"),
    fig_size=(3000, 750),
    col_width=[1, 3.4, 6, 2, 2.7,2.5]
)

df2img.save_dataframe(fig=fig, filename="df_arrival_head.png")
```

```python
df_assoc = pd.read_csv('/wd1/hbito_data/data/datasets_all_regions/assoc_2010_2015_reloc_cog_ver3.csv',index_col=0)
df_assoc.slatitude = df_assoc.slatitude.round(2)
df_assoc.slongitude = df_assoc.slongitude.round(2)
df_assoc.selevation = df_assoc.selevation.round(2)
df_assoc.delta = df_assoc.delta.round(2)
df_assoc.esaz = df_assoc.esaz.round(2)
df_assoc.seaz = df_assoc.seaz.round(2)

df_assoc.head()
```

```python
fig = df2img.plot_dataframe(
    df_assoc.head(),

    tbl_header=dict(
        align="center",
        fill_color="purple",
        font_color="white",
        font_size=80,
        line_color="darkslategray",
        height=120
    ),
    tbl_cells=dict(
        align="center",
        line_color="darkslategray",
        font_size=90,
        height=120
    ),
    row_fill_color=("#ffffff", "#d7d8d6"),
    fig_size=(5800, 750),
    col_width=[0.5, 0.5,0.5, 1.0, 0.7,0.7,0.7,1,1,1,0.9,0.9,0.9]
)

df2img.save_dataframe(fig=fig, filename="/wd1/hbito_data/data/datasets_all_regions/df_assoc_head.png")
```

```python
df_origin = pd.read_csv('/wd1/hbito_data/data/datasets_all_regions/origin_2010_2015_reloc_cog_ver3.csv',index_col=0)
df_origin.lat = df_origin.lat.round(2)
df_origin.lon = df_origin.lon.round(2)
df_origin.depth = df_origin.depth.round(2)
df_origin.gap = df_origin.gap.round(2)
df_origin.head()
```
