```python
from pnwstore.mseed import WaveformClient
```

```python
from obspy.core.utcdatetime import UTCDateTime
```

```python
client = WaveformClient()
```

```python
s = client.get_waveforms(network="7D", station="FS15D", channel="HH?",year=2014,day=20,month=9)
s
```

```python
s.filter(type='bandpass',freqmin=4,freqmax=15)
```

```python
s[1].plot(type='dayplot');
```

```python
s[2].plot(type='dayplot');
```

```python
s.plot();
```

```python
dt = UTCDateTime("2014-08-12T13:30:00.014100Z")
```

```python
s.trim(dt, dt+30)
```

```python
s.plot()
```
