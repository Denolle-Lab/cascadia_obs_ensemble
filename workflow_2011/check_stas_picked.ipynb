## Checking if the catalogs for all the stations were produced for 2011-2015

```python
import os
import glob
from obspy.clients.fdsn.client import Client
import datetime
import pandas as pd
import numpy as np
```

### 1. Check stations for 2011

```python
# Define parameters 
year1 = 2011# Change the working directory
os.chdir(f'/home/hbito/cascadia_obs_ensemble/data/picks_{year}')
print(os.getcwd())
```

```python
# Make an array of stations and their networks that we made in 1_parallel_detect_picks_elep_2011
# The code below is similar to the first part in 1_parallel_detect_picks_elep_2011 to make the station array

# Define clients
client_inventory = Client('IRIS')

# Now create your list of days to loop over!
time1 = datetime.datetime(year=year1,month=1,day=1)
time2 = datetime.datetime(year=year1+1,month=1,day=1)
time_bins = pd.to_datetime(np.arange(time1,time2,pd.Timedelta(1,'days')))

inventory = client_inventory.get_stations(network="C8,7D,7A,CN,NV,UW,UO,NC,BK,TA,OO,PB,X6,Z5,X9", 
                                          station="*", minlatitude=40,minlongitude=-127,maxlatitude=50,maxlongitude=-123, starttime=time1.strftime('%Y%m%d'),endtime=time2.strftime('%Y%m%d'))

# Make a list of networks and stations
networks_stas = []

for i in range(len(inventory)):
    network = inventory[i].code
    
    for j in range(len(inventory[i])):
        networks_stas.append([network,inventory[i].stations[j].code])
    
networks_stas1 =np.array(networks_stas)
networks_stas1
```

```python
# ls_sta = ['I02D', 'J01E', 'FS16B', 'J33B', 'M14B', 'FS20B', 'G26B',
#           'M09B', 'FS17B', 'G27B', 'M12B', 'FS13B', 'J09B', 'G17B', 'J25B',
#           'G19B', 'FS09B', 'G10B', 'G25B', 'BABR', 'KBO', 'KOM', 'KRMB',
#           'KHMB', 'KHBB', 'KSXB', 'KEB', 'KMPB', 'JCC']
available_stas = []
unavailable_stas = []
for i in range(len(networks_stas1)):
    network =networks_stas1[i][0]
    station = networks_stas1[i][1]
    file_pattern = f"{station}_*.csv"
    matching_files = glob.glob(file_pattern)
    # Sort the files by the date part of the filename
    matching_files_sorted = sorted(matching_files, key=lambda x: x.split('_')[1])
    if matching_files_sorted:
        print(f"Catalogs for {network,station}:")
        for file in matching_files_sorted:
            print(file)
        available_stas.append([network,station])
    else:
        unavailable_stas.append([network,station])
        print(f"No catalogs found for {network,station}")
```
